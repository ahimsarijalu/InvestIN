###Mencoba cleaning pake IQR

```python
df_temp = df.copy()
Q1 = df_temp.quantile(0.25)
Q3 = df_temp.quantile(0.75)

IQR = Q3 - Q1
print(IQR)
```

```python
df_clean = df_temp.copy()
no = 0
for col in df_clean.columns:
  df_clean = df_clean[df_clean[col] > (Q1[no] - 1.5 * IQR[no])]
  df_clean = df_clean[df_clean[col] < (Q3[no] + 1.5 * IQR[no])]
  no += 1
```

## Data Normalization

MinMax Normalization

```python
df_min_max = df_clean.copy()
  
# minmax normalization 
for column in df_min_max.columns:
    df_min_max[column] = (df_min_max[column] - df_min_max[column].min()) / (df_min_max[column].max() - df_min_max[column].min())    
  
print(df_min_max)
```

```python
# before minmax
df.plot(kind='bar')
```

```python
# minmax normalized
df_min_max.plot(kind = 'bar')
```

Z-score Normalization

lebih baik untuk menangani outlier

```python
df_z_score= df_clean.copy()
  
# z-score normalization 
for column in df_z_score.columns:
    df_z_score[column] = (df_z_score[column] -
                           df_z_score[column].mean()) / df_z_score[column].std()    
   
display(df_z_score)
```

```python
from scipy import stats

df_zscore = df.copy()
z = np.abs(stats.zscore(df_zscore))
print(z)
```
